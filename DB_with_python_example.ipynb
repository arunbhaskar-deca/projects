```python
from mysql import connector as MSC
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine
```

```python
connection = MSC.connect(user='root',password='college',host='localhost',database='employees', ssl_disabled=True)

cursor = connection.cursor()
```

```python
query = """
SELECT * FROM employees e
JOIN salaries s
ON e.emp_no=s.emp_no
WHERE e.hire_date > '1999-01-01';
"""
```

```python
cursor.execute(query)
```

```python
results = []
for i, data in enumerate(cursor):
    results.append(data)
cursor.close()
connection.close()
```

```python
df=pd.DataFrame(results)
df.columns = ['emp_no',
'birth_date',
'first_name',
'last_name',
'gender',
'hire_date',
'emp_no_2',
'salary',
'from_date',
'to_date']
df.head()
```
